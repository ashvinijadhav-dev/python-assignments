**1. Write a function is_anagram (s1, s2) that checks whether two strings are anagrams.**

```python
def is_anagram(s1, s2):
    s1 = s1.lower().replace(" ", "")
    s2 = s2.lower().replace(" ", "")

    if sorted(s1) == sorted(s2):
        print("Anagram")
    else:
        print("Not Anagram")

is_anagram("listen", "silent")
```

**3. Write a function intersection(list1, list2) that returns common elements between two lists without duplicates.**

```python
def intersection(list1, list2):
    common = []

    for i in list1:
        if i in list2 and i not in common:
            common = common + [i]

    return common


result = intersection([7,6,8,4,5,3], [5,4,3,2,1,9])
print(result)
```

**5. Write a function that takes a list of integers and a target value. Return the indices of two**
**numbers whose sum equals the target.**

```python
def find_indices(numbers, target):
    for i in range(len(numbers)):
        for j in range(i + 1, len(numbers)):
            if numbers[i] + numbers[j] == target:
                return i, j


result = find_indices([2, 7, 11, 15], 9)
print(result)
```

**1. Write a function generate_password(length=8) that generates a random secure password of given length.**

```python
import random
import string

def generate_password(length=8):
    characters = string.ascii_letters + string.digits + string.punctuation

    password = ""

    for i in range(length):
        password += random.choice(characters)

    return password


print(generate_password())
print(generate_password(12))

```

**5. Write a function validate_password (password, min_length=8) that checks password rules**
**(uppercase, lowercase, number, special character).**

```python
def validate_password(password, min_length=8):
    upper = lower = number = special = 0

    for ch in password:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1
        elif ch.isdigit():
            number += 1
        else:
            special += 1

    if len(password) >= min_length and upper > 0 and lower > 0 and number > 0 and special > 0:
        print("Valid Password")
    else:
        print("Invalid Password")


validate_password("Python@123")
validate_password("python123")
```

# **3) Recursion (Factorial-type Problems)**

**1. Write a recursive function to calculate the factorial of a number.**

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

**2. Write a recursive function to reverse a string.**

```python
def reverse_string(s):
    if s == "":
        return ""
    return reverse_string(s[1:]) + s[0]

print(reverse_string("Python"))
```

**3. Write a recursive function to find the nth Fibonacci number.**

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(6))
```

**4. Write a recursive function to convert a decimal number to binary**

```python
def decimal_to_binary(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return str(decimal_to_binary(n // 2)) + str(n % 2)

print(decimal_to_binary(10))
```

**5. Write a recursive function to check whether a string is palindrome.**

```python
def palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return palindrome(s[1:-1])

word = "madam"

if palindrome(word):
    print("Palindrome")
else:
    print("Not Palindrome")
```

**1. Write a function that accepts any number of arguments using *args and returns their sum.***

```python
def add(*args):
    total = 0
    for i in args:
        total += i
    return total

print(add(10, 20, 30, 40))
```

**2. Write a function using *args that returns all prime numbers among the arguments.***

```python
def prime_numbers(*args):
    for n in args:
        if n > 1:
            count = 0
            for i in range(1, n + 1):
                if n % i == 0:
                    count += 1
            if count == 2:
                print(n)

prime_numbers(2, 4, 5, 8, 11, 15, 17)
```

```python
def student_grade(*args, **kwargs):
    total = 0

    for mark in args:
        total += mark

    average = total / len(args)

    print("Name :", kwargs["name"])
    print("Course :", kwargs["course"])
    print("Average :", average)

    if average >= 90:
        print("Grade : A")
    elif average >= 75:
        print("Grade : B")
    elif average >= 50:
        print("Grade : C")
    else:
        print("Grade : Fail")

student_grade(80, 90, 70, name="Ashvini", course="Python")
```

**2. Write a function that returns the longest word in a sentence.**

```python
def longest_word(sentence):
    words = sentence.split()
    longest = words[0]

    for word in words:
        if len(word) > len(longest):
            longest = word

    return longest

print(longest_word("I am learning Python Programming"))
```

**3. Write a function that returns True if a number is an Armstrong number, otherwise False.**

```python
def armstrong(n):
    temp = n
    total = 0

    while temp > 0:
        digit = temp % 10
        total = total + digit ** 3
        temp = temp // 10

    return total == n

print(armstrong(153))
print(armstrong(123))
```

**4. Write a function that returns a list of numbers divisible by both 3 and 5 between two numbers.**

```python
def divisible(start, end):
    result = []

    for i in range(start, end + 1):
        if i % 3 == 0 and i % 5 == 0:
            result.append(i)

    return result

print(divisible(1, 100))
```

**5. Write a function that returns the number of uppercase, lowercase, and digits in a string.**

```python
def count_characters(text):
    upper = 0
    lower = 0
    digit = 0

    for ch in text:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1
        elif ch.isdigit():
            digit += 1

    return upper, lower, digit

u, l, d = count_characters("Python123ABC")

print("Uppercase:", u)
print("Lowercase:", l)
print("Digits:", d)
```

**1. Write a program demonstrating the difference between global and local variables.**

```python
x = 10      # Global variable

def demo():
    y = 20  # Local variable
    print("Inside function")
    print("Global:", x)
    print("Local:", y)

demo()

print("Outside function")
print("Global:", x)
```
