***Q2. [EASY] Palindrome Number***

***Given an integer n, return True if it reads the same from left to right and right to left, otherwise return False.***


```python
# Example 1
def is_palindrome(n):
    original_num = n
    reverse_num = 0
    
    while n > 0:
     digit = n % 10
     reverse_num = reverse_num * 10 + digit
     n = n // 10

    return original_num == reverse_num

is_palindrome(121)
        
    
```

```python
# Example 2
def is_palindrome1(n):
    original = n
    reverse = 0
    
    while n > 0:
     digit = n % 10
     reverse = reverse * 10 + digit
     n = n // 10

    return original == reverse

is_palindrome1(123)
        
    
```

***Q15. [MEDIUM] Percentage Calculator Using Functions***

***A student scores marks in 5 subjects (each out of 100). Write TWO functions:***
***calculate_percentage(marks) that returns the percentage, and assign_grade(percentage) that returns a grade based on the percentage.***

```python
def calculate_percentage(marks):
    total = sum(marks)
    percentage = (total / 500) * 100
    return total, percentage


def assign_grade(percentage):
    if percentage >= 90:
        return "A+"
    elif percentage >= 80:
        return "A"
    elif percentage >= 70:
        return "B"
    elif percentage >= 60:
        return "C"
    elif percentage >= 50:
        return "D"
    else:
        return "Fail"


# Input marks for 5 subjects
marks = []

for i in range(5):
    mark = int(input(f"Enter marks of subject {i + 1}: "))
    marks.append(mark)

# Function calls
total, percentage = calculate_percentage(marks)
grade = assign_grade(percentage)

# Output
print("Total Marks =", total, "/ 500")
print("Percentage =", percentage, "%")
print("Grade =", grade)
```

**Q29. [HARD] Student Marks Report Card**

**Write a function student_report(maths, science, english) that takes marks in 3 subjects and returns the**
**overall result using nested if/elif/else. If any subject is below 35, the student fails.**

```python
# example_1
def student_report(maths, science, english):
    if maths < 35 or science < 35 or english < 35:
        return "Fail"

    average = (maths + science + english) / 3

    if average >= 75:
        return "Distinction"
    elif average >= 60:
        return "First Class"
    elif average >= 50:
        return "Second Class"
    else:
        return "Pass Class"

print(student_report(95, 92, 90))
```
