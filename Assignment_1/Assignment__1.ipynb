# 8. Simple Interest

***Question:Write a Python program that accepts principal amount, rate of interest, and time period and calculates simple interest.***

***SI= P×R×T/100***




```python
principal_amount = float(input('Enter principal amount :'))
rate_of_interest = float(input('Enter rate of interest :'))
time_period = float(input('Enter time period :'))
simple_interest = (principal_amount * rate_of_interest * time_period) / 100
simple_interest
```

# 9. Total Amount Using Simple Interest

***Question:Write a Python program that calculates the final amount after adding simple interest to the principal.***

***formula : Amount=P+SI***


```python
principal_amount = float(input('Enter principal amount :'))
rate_of_interest = float(input('Enter rate of interest :'))
time_period = float(input('Enter time period :'))

simple_interest = (principal_amount * rate_of_interest * time_period) / 100
Total_amount =principal_amount + simple_interest

print("Simple interest =", simple_interest)
print("Total amount =", Total_amount )
```

# 18. Distance Between Two Points

***Question:Write a Python program that accepts coordinates ((x_1,y_1)) and ((x_2,y_2)) and calculates the distance between them.***


```python
import math

x1 = float(input("Enter x1: "))
y1 = float(input("Enter y1: "))
x2 = float(input("Enter x2: "))
y2 = float(input("Enter y2: "))

distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

print("Distance between the two points =", distance)
```

# 20. Body Mass Index (BMI)

***Write a Python program that accepts a person's weight (in kilograms) and height (in meters) and calculates their Body Mass Index (BMI).***

***formula : BMI = Weight ÷ (Height × Height)***

```python
weight = float(input("Enter your weight in kilograms: "))
height = float(input("Enter your height in meters: "))

bmi = weight / (height ** 2)

print("Your BMI is =", bmi)
```
