**1. Ramesh’s basic salary is input through the keyboard. His dearness allowance is 40% of basic salary, and
house rent allowance is 20% of basic salary. Write a program to calculate his gross salary.**

```python
basic_salary = float(input('enter your basic salary:'))

dearness_allowance = basic_salary * 40/100

house_rent_allowance = basic_salary * 20/100

Gross_Salary = (basic_salary + dearness_allowance  + house_rent_allowance)

print(f'dearness_allowance: {dearness_allowance}, house_rent_allowance : {house_rent_allowance}')
print('Ramesh’s Gross salary is:',Gross_Salary)
```

***4. Temperature of a city in Fahrenheit degrees is input through the keyboard. Write a program to convert this
temperature into Centigrade degrees.***

```python
fahrenheit = float(input('Enter temperature of a city in Fahrenheit: '))

centigrade = (fahrenheit - 32) * 5 / 9

print('Temperature of a city in Centigrade =', round(centigrade,2))
```

***9. In a town, the percentage of men is 52. The percentage of total literacy is 48. If total percentage of literate
men is 35 of the total population, write a program to find the total number of illiterate men and women if
the population of the town is 80,000.***

```python
total_population = 80000

pct_men = 52
pct_total_literacy = 48
pct_lit_men = 35

total_men = (pct_men / 100) * total_population
total_women = total_population - total_men

total_literate = (pct_total_literacy / 100) * total_population
lit_men = (pct_lit_men / 100) * total_population


illit_men = total_men - lit_men
lit_women = total_literate - lit_men
illit_women = total_women - lit_women


print(f"Total Population: {total_population}")
print(f"Total Number of Illiterate Men: {int(illit_men)}")
print(f"Total Number of Illiterate Women: {int(illit_women)}")
```

***10. A cashier has currency notes of denominations 10, 50 and 100. If the amount to be withdrawn is input
through the keyboard in hundreds, find the total number of currency notes of each denomination the cashier will have to give to the withdrawer.***

```python

amount = int(input("Enter the amount to be withdrawn : "))

notes_100 = amount // 100
remaining_amount = amount % 100
 
notes_50 = remaining_amount // 50
remaining_amount = remaining_amount % 50

notes_10 = remaining_amount // 10
remaining_amount = remaining_amount % 10


print(f"\nCurrency Breakdown for Rs.: {amount}")
print(f"Rs. 100 notes: {notes_100}")
print(f"Rs. 50 notes : {notes_50}")
print(f"Rs. 10 notes : {notes_10}")


if remaining_amount > 0:
    print(f"Remaining change that cannot be given in available denominations: Rs. {remaining_amount}")
```
