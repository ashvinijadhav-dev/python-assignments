***1. wap ask the user take two number print subtraction multiplication and division***

Example: replace a and b with random integers using random package


```python
import random

a=random.randint(1,100)
b=random.randint(1,100)
c=a+b

print("a=",a)
print("b=",b)

print("addition=",c)
print("Subtraction=",a-b)
print("Multiplication=",a*b)
print("Division =", round(a / b, 2))
```

***2.wap ask the user take the radius find the area of circle***

Example: take pi value from math package

r value using random

```python
import math
import random

pi=math.pi
r=random.randint(1,100)

area_of_circle=pi*r*r

print('area_of_circle :',round(area_of_circle,2))
```
