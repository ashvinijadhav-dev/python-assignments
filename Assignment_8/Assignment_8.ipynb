# 2.Ask the user enter a number get it is an even or odd
**repeat the above task 5 times**


```python
for i in range(5):
    num = int(input("Enter a number: "))

    if num % 2 == 0:
        print("Even Number")
    else:
        print("Odd Number")
```

# 3.Ask the user print 7th table   range 1 to 11      
**7x1=7**
**7x2=14**

```python
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

# 5.Ask the user get a random number between 1 to 10
**user enter a number between 1 to 10**

**if both are match print won otheriwse lost**

**give only 3 chances**

```python
import random

random_number = random.randint(1, 10)

for i in range(3):
    user_number = int(input("Enter a number between 1 and 10: "))

    if user_number == random_number:
        print("You Won!")
        break
    else:
        print("You Lost!")

print("Random Number was:", random_number)
```
