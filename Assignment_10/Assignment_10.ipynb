 **mutable vs immutable**

```python
list1 = ['w','e','l','c','o','m','e']

print("Before:", list1)

list1[0] = 'W'

print("After:", list1)
```

```python
list2 = [10,20,30,40,50,60]

print("Before:", list2)

list2[2] = 300

print("After:", list2)
```

```python
list3 = ['w','e','l','c','o','m','e',10,20,30,40,50,60]

print("Before:", list3)

list3[7] = 100

print("After:", list3)
```
